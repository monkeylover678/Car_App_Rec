```python
import re

def normalize_color(color):
    color = color.lower()
    if re.search(r'black|ebony|graphite|carbon|nero|obsidian', color):
        return 'Black'
    elif re.search(r'white|ivory|pearl|ice|blizzard|snow|bianco|wht|whi|whitr|whc|aspen', color):
        return 'White'
    elif re.search(r'blue|aqua|cyan|navy|azure|ocean|sky|cobalt|blu|deep sea', color):
        return 'Blue'
    elif re.search(r'red|crimson|burgundy|scarlet|rosso|ruby|garnet|hot tamale', color):
        return 'Red'
    elif re.search(r'silver|gray|grey|mercury|platinum|metallic|steel|pewter|smokey|tungsten|chrome|shimmering', color):
        return 'Silver'
    elif re.search(r'green|emerald|olive|verde|cactus|moss|teal', color):
        return 'Green'
    elif re.search(r'yellow|gold|champagne|citrine|saffron|giallo', color):
        return 'Yellow'
    elif re.search(r'brown|bronze|tan|beige|mocha|coffee|copper|maroon|cashmere', color):
        return 'Brown'
    elif re.search(r'purple|violet|lavender|mauve|amethyst|viola', color):
        return 'Purple'
    elif re.search(r'pink|rose|fuchsia|eggplant', color):
        return 'Pink'
    elif re.search(r'orange|tangerine|lava|habanero|mango', color):
        return 'Orange'
    elif re.search(r'caviar|night|shadow|charcoal|dark|blackout|obsidian', color):
        return 'Black'
    elif re.search(r'special|custom|other|not specified|none|void', color):
        return 'Other'
    else:
        return 'Other'

df['Exterior color'] = df['Exterior color'].apply(normalize_color)
df['Interior color'] = df['Interior color'].apply(normalize_color)
df.info()
```

```python
# Normalize Drivetrain Column
def normalize_drivetrain(drivetrain):
    drivetrain = drivetrain.lower()
    if drivetrain in ['rear-wheel drive', 'rear wheel drive']:
        return 'RWD'
    elif drivetrain in ['all-wheel drive', 'all wheel drive', 'awd']:
        return 'AWD'
    elif drivetrain in ['four-wheel drive', 'four wheel drive', '4wd']:
        return '4WD'
    elif drivetrain in ['front-wheel drive', 'front wheel drive', 'fwd']:
        return 'FWD'
    else:
        return 'Unknown'

df['Drivetrain'] = df['Drivetrain'].apply(normalize_drivetrain)
df['Drivetrain'].unique()
```

```python
# Normalize Fuel Type Column
def normalize_fuel_type(fuel_type):
    fuel_type = fuel_type.lower()
    if 'gasoline' in fuel_type or 'regular unleaded' in fuel_type:
        return 'Gasoline'
    elif 'diesel' in fuel_type:
        return 'Diesel'
    elif 'hybrid' in fuel_type:
        return 'Hybrid'
    elif 'electric' in fuel_type or 'phev' in fuel_type or 'plug-in' in fuel_type:
        return 'Electric'
    elif 'compressed natural gas' in fuel_type or 'natural gas' in fuel_type:
        return 'CNG'
    elif 'flex fuel' in fuel_type or 'e85' in fuel_type:
        return 'Flex Fuel'
    elif 'premium unleaded' in fuel_type:
        return 'Premium'
    else:
        return 'Other'

df['Fuel type'] = df['Fuel type'].apply(normalize_fuel_type)
df['Fuel type'].unique()
```

```python
# Normalize Transmission Column
def normalize_transmission(transmission):
    transmission = transmission.lower()
    if 'manual' in transmission:
        return 'Manual'
    elif 'automatic' in transmission or 'a/t' in transmission or 'auto' in transmission:
        return 'Automatic'
    elif 'cvt' in transmission:
        return 'CVT'
    else:
        return transmission

df['Transmission'] = df['Transmission'].apply(normalize_transmission)
df['Transmission'].unique()
```

```python
# Normalize Engine Column to Categorized Types
def normalize_engine(engine):
    engine = engine.lower()
    if re.search(r'v8', engine):
        return 'V8'
    elif re.search(r'v6', engine):
        return 'V6'
    elif re.search(r'v12', engine):
        return 'V12'
    elif re.search(r'v10', engine):
        return 'V10'
    elif re.search(r'i4|inline-4|4-cylinder|1\.4|1\.5|1\.6|2\.0|2\.4', engine):
        return '4-Cylinder'
    elif re.search(r'3-cylinder|i3|inline-3|1\.0|1\.2|1\.3', engine):
        return '3-Cylinder'
    elif re.search(r'5-cylinder|i5|inline-5|2\.5', engine):
        return '5-Cylinder'
    elif re.search(r'6-cylinder|i6|inline-6|3\.0|3\.2|3\.5|3\.6', engine):
        return '6-Cylinder'
    elif re.search(r'turbo', engine):
        return 'Turbo'
    elif re.search(r'supercharged', engine):
        return 'Supercharged'
    elif re.search(r'ev|electric|phev|hybrid', engine):
        return 'Electric/Hybrid'
    else:
        match = re.search(r'(\d+\.?\d*)[lL]?', engine)
        if match:
            engine = match.group(1) + 'L'
            engine_capacity = float(match.group(1))
            if engine_capacity > 8.0:
                engine = 'Other'
            else:
                engine = engine
        else:
            engine = 'Other'
        # Remove additional information after engine capacity
        engine = re.sub(r'(\d+\.?\d*)[lL]?.*', r'\1L', engine)
        return engine
        
df['Engine'] = df['Engine'].apply(normalize_engine)
df['Engine'].nunique()
```

```python
unique_engines = []
for key, val in zip(df['Engine'].sort_values().value_counts().index, df['Engine'].sort_values().value_counts()):
    unique_engines.append((key, val))
unique_engines = tuple(unique_engines)
unique_engines
```

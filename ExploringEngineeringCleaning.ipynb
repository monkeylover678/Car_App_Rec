```python
import pandas as pd

# Extract the 'Title' column into a variable
data = df['Title']

# Predefined list of multi-word makes
multi_word_makes = [
    "Alfa Romeo", "Aston Martin", "Land Rover", "Am General"
]

# Function to split Original Data into Year, Make, and Model
def split_data(row):
    parts = row.split(' ', 3)  # Split into up to 4 parts
    if len(parts) < 3:
        return pd.Series([None, None, None])  # Return nulls if format is unexpected
    
    year = parts[0] if parts[0].isdigit() else None
    make = None
    model = None

    # Check for multi-word makes
    for mw_make in multi_word_makes:
        if row.startswith(year + ' ' + mw_make):
            make = mw_make
            model = row[len(year) + len(mw_make) + 2:].strip()  # Extract model
            break
    
    # Handle single-word makes
    if make is None:
        make = parts[1]
        model = ' '.join(parts[2:])
    
    return pd.Series([year, make, model])

# Apply the function to the 'data' series
split_results = data.apply(split_data)

# Assign the resulting columns back to the original DataFrame
df[['Year', 'Make', 'Model']] = split_results

# Display the updated DataFrame
df
```

```python
# Function to clean the mileage
def clean_mileage(mileage):
    # Remove commas and the "mi." suffix, then convert to integer
    return int(mileage.replace(',', '').replace(' mi.', '').strip())

# Apply the function to the Mileage column
df['Mileage'] = df['Mileage'].apply(clean_mileage)
df
```

```python
# Function to clean the mileage
def clean_price(price):
    # Remove commas and the "mi." suffix, then convert to integer
    return int(price.replace(',', '').replace('$', '').strip())

# Apply the function to the Mileage column
df['Price(USD)'] = df['Price(USD)'].apply(clean_price)
df
```

```python
# Function to clean the 'Condition' column
def normalize_state(value):
    # Define acceptable values
    valid_conditions = ['used', 'new', 'certified']
    # Check if any valid condition exists in the value
    for condition in valid_conditions:
        if condition in value.lower():
            return condition
    # If no valid condition is found, return None
    return None

# Apply the function to clean the column
df['State'] = df['State'].apply(normalize_state)

# Display the cleaned DataFrame
df
```

```python
import pandas as pd
import json
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter

# Example DataFrame with JSON strings
data = pd.DataFrame(df['Basics'])

# Function to extract keys from JSON text
def extract_keys(json_text):
    try:
        json_obj = json.loads(json_text)  # Parse JSON string into a dictionary
        return list(json_obj.keys())  # Return the keys of the JSON object
    except json.JSONDecodeError:
        return []  # Return an empty list if the JSON is invalid

# Extract keys from each JSON text and flatten the list
all_keys = []
data['Basics'].apply(lambda x: all_keys.extend(extract_keys(x)))

# Count the frequency of each key
key_counts = Counter(all_keys)

# Generate a word cloud|
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(key_counts)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('JSON Key Trends')
plt.show()
```

```python
# Function to extract a value for a specific key from JSON text
def extract_value(json_text, key):
    try:
        json_obj = json.loads(json_text)
        return json_obj.get(key, None)
    except json.JSONDecodeError:
        return None

# Step 1: Determine the frequency of keys across all rows
all_keys = []
df['Basics'].apply(lambda x: all_keys.extend(extract_keys(x)))
key_counts = Counter(all_keys)

# Step 2: Select frequent keys (for example, those that appear at least twice)
frequent_keys = [key for key, count in key_counts.items() if count >= 2 and key != "Mileage"]

# Step 3: Create new columns for each frequent key
for key in frequent_keys:
    df[key] = df['Basics'].apply(lambda x: extract_value(x, key))

# Display the updated DataFrame
df
```
